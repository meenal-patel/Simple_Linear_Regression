```python
import sklearn
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import statsmodels.formula.api as smf
from sklearn.metrics import r2_score
```

```python
chinagdp = pd.read_csv("china_gdp.csv")

print (chinagdp.shape,"china_gdp.csv")
print ("="*30)

print (chinagdp.head(10))

plt.figure(figsize=(20,7),facecolor="lightgreen")
plt.scatter(chinagdp.Year,chinagdp.Value)
plt.xlabel("Year");
plt.ylabel("Value")
# print (china_gdp.AT)
plt.show()
```

```python
chinagdp.corr()
```

```python
lm = LinearRegression()

X = chinagdp["Year"].values.reshape(-1,1)
Y = chinagdp["Value"].values.reshape(-1,1)

lm.fit(X,Y)
```

```python
plt.figure(figsize=(10,7),facecolor="lightgreen")
plt.scatter(chinagdp.Year,(chinagdp.Value),color="black",label="actual");
plt.plot(chinagdp.Year,lm.predict(X),label="predicted",color="green",linestyle="-.");
plt.scatter(chinagdp.Year,lm.predict(X),color="red")
plt.xlabel("Year");
plt.ylabel("Value")
#plt.xlim(1950,2015)
#plt.ylim()
# print (wcat.AT)
plt.legend(loc="best")
plt.show()

```

```python
plt.figure(figsize=(10,7),facecolor="lightgreen")
plt.scatter(chinagdp.Year,chinagdp.Value,color="black",label="actual");
plt.plot(chinagdp.Year,first_model.predict(),label="predicted",color="red",linestyle="-.");
plt.scatter(chinagdp.Year,first_model.predict(),color="red")
plt.xlabel("Year");
plt.ylabel("Value")
# print (wcat.AT)
plt.legend(loc="best")
plt.show()

```

```python
second_model=smf.ols("Value~np.log(Year)",data=chinagdp).fit()
```

```python
plt.figure(figsize=(10,7),facecolor="lightgreen")
plt.scatter(chinagdp.Year,chinagdp.Value,color="black",label="actual");
plt.plot(chinagdp.Year,second_model.predict(),label="predicted",color="red",linestyle="-.");
plt.scatter(chinagdp.Year,second_model.predict(),color="red")
plt.xlabel("Year");
plt.ylabel("Value")
# print (wcat.AT)
plt.legend(loc="best")
plt.show()
```

```python
second_model.summary()
```

```python
third_model=smf.ols("np.log(Value)~(Year)",data=chinagdp).fit()
```

```python
third_model.summary()
```

```python
plt.figure(figsize=(10,7),facecolor="lightgreen")
plt.scatter(chinagdp.Year,np.log(chinagdp.Value),color="black",label="actual");
plt.plot(chinagdp.Year,third_model.predict(),label="predicted",color="red",linestyle="-.");
plt.scatter(chinagdp.Year,third_model.predict(),color="red")
plt.xlabel("Year");
plt.ylabel("Value")
# print (wcat.AT)
plt.legend(loc="best")
plt.show()
```

```python
# Third model is best bcoz is has higher R2 value
```
